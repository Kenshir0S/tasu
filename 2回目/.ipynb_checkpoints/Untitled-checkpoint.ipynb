```python
import pandas as pd
import json
from pandas import json_normalize
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
db_file_name = "datas.json"
```

```python
with open(db_file_name, "r") as f:
        json_datas = json.load(f)
df = json_normalize(json_datas)
```

```python
df
```

```python
sns.barplot(x="date", y="price", hue="in_out", data=df, ci=None)
plt.show()
```

```python
import altair as alt
```

```python
df = df.sort_values("date")
```

```python
df = df.pivot_table(index="date", columns="category", values="price")
```

```python
df
```

```python
df = df.apply(lambda x: x.sum(), axis=1)
```

```python
df = pd.DataFrame(df, columns=["price"]).reset_index()
```

```python
df
```

```python
alt.Chart(df).mark_line(clip=True).encode(
    x='date:T',
    y='price:Q',
    color='category:N',
    # strokeDash='category',
).interactive()
```
